```python
# France

size = (106,190)
imgFranceBlue = Image.new('RGB',size,"blue")
imgFranceWhite = Image.new('RGB',size,"white")
imgFranceRed = Image.new('RGB',size,"red")
 
imgFrance = Image.new('RGB', (318,190))
 
imgFrance.paste(imgFranceBlue, (0,0))
imgFrance.paste(imgFranceWhite, (106,0))
imgFrance.paste(imgFranceRed, (212,0))
 
imgFrance = ImageOps.expand(imgFrance, border=1, fill='black')
imgFrance.save("france", "PNG")
imgFrance.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def computeHistogram(image):
    histogram, _ = np.histogram(image.flatten(), bins=256, range=[0,256])
    return histogram

def computeCumulativeHistogram(histogram):
    cumulative_histogram = np.cumsum(histogram)
    return cumulative_histogram

def normalizeImage(cumulative_histogram, total_pixels):

    normalized_histogram = cumulative_histogram / total_pixels
    return normalized_histogram

def histEqualization(image):
    histogram = computeHistogram(image)
    cumulative_histogram = computeCumulativeHistogram(histogram)
    total_pixels = image.size
    normalized_histogram = normalizeImage(cumulative_histogram, total_pixels)
    equalized_image = np.interp(image.flatten(), range(256), normalized_histogram * 255)
    equalized_image = equalized_image.reshape(image.shape).astype(np.uint8)
    return equalized_image
```

```python
myimage = np.array(Image.open('image2.png'))

w,h=400,400
plt.figure(figsize=(w/80,h/80))  # optional, to control the size of figure (unit: pixel)


plt.axis('off') # optional: remove axes
#plt.hist(myimage)
plt.imshow(myimage)
plt.show()

myimage = histEqualization(myimage)


w,h=400,400
plt.figure(figsize=(w/80,h/80))  # optional, to control the size of figure (unit: pixel)


plt.axis('off') # optional: remove axes
#plt.hist(myimage)
plt.imshow(myimage)
plt.show()
```

```python
def getColorScheme(img,k):
    t=np.zeros((len(img),len(img[0]))
    for i in range(len(img)):
        for j in range(len(img[i])):
            t[i][j]=img[i][j][k]
    return t

imgR = getColorScheme(img1.tolist(),0) 
plt.imshow(imgR)
plt.show()
imgG = getColorScheme(img1,1) 
plt.imshow(imgG)
plt.show()
imgB = getColorScheme(img1,2) 
plt.imshow(imgB)
plt.show()

```
