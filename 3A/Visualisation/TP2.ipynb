### Question 1 :

```python
# France

size = (106,190)
imgFranceBlue = Image.new('RGB',size,"blue")
imgFranceWhite = Image.new('RGB',size,"white")
imgFranceRed = Image.new('RGB',size,"red")
 
imgFrance = Image.new('RGB', (318,190))
 
imgFrance.paste(imgFranceBlue, (0,0))
imgFrance.paste(imgFranceWhite, (106,0))
imgFrance.paste(imgFranceRed, (212,0))
 
imgFrance = ImageOps.expand(imgFrance, border=1, fill='black')
imgFrance.save("france", "PNG")
imgFrance.show()
```

```python
# Italie

size = (106,190)
imgItalieGreen = Image.new('RGB',size,"green")
imgItalieWhite = Image.new('RGB',size,"white")
imgItalieRed = Image.new('RGB',size,"red")
 
imgItalie = Image.new('RGB', (318,190))
 
imgItalie.paste(imgItalieGreen, (0,0))
imgItalie.paste(imgItalieWhite, (106,0))
imgItalie.paste(imgItalieRed, (212,0))
 
imgItalie = ImageOps.expand(imgItalie, border=1, fill='black')
imgItalie.save("italie", "PNG")
imgItalie.show()
```

### Question 6 :

```python
def assombrir(img):
    img = img.point(lambda p: (p/2))
    return img

assombrir(Image.open('image1.jpg')).show()
```

```python
def eclaircir(img):
    img = img.point(lambda p: (p*2))
    return img

eclaircir(Image.open('image1.jpg')).show()
```
