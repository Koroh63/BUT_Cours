```python
img1 = Image.open('lena.png')
img2 = Image.open('mandrill.png')
print(img1)
def fonduImage(img1,img2,ratio):
    final = np.zeros((len(img1),len(img1[0])))
    for i in range(len(img1)):
        for j in range(len(img1[0])):
            final[i][j]=img1[i][j]*ratio+img2[i][j]*(1-ratio)           
            
    
    return final
                       
fondu = fonduImage(np.array(img1),np.array(img2),0.8)
print(fondu)
plt.imshow(fondu)
plt.show()
```
