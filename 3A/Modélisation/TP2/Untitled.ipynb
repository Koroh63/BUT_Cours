```python
mat1 = np.random.randint(0,15,(5,6))
print(mat1)
```

## Questiion 2 :

```python
for col in range(1,mat1.shape[1]-1,2):
    mat1[:,col]=mat1[:,col]-2 * mat1[:,col+1]
print(mat1)
```

```python
mask = mat1<0
print(mask)
mat1[mask] = 0
print(mat1)
```
